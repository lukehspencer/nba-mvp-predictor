```python
# Replace a player's team with "TOT" if they played for multiple teams in a season

players["Team"].replace({"2TM": "TOT"}, inplace=True)
players["Team"].replace({"3TM": "TOT"}, inplace=True)
players["Team"].replace({"4TM": "TOT"}, inplace=True)
players["Team"].replace({"5TM": "TOT"}, inplace=True)
```

```python
players.groupby(["Player", "Year"]).get_group(("Bobby Jones", 2008))
```

```python
# Makes sure that each player only has a single row for each year

def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Team"]=="TOT"]
        row["Team"] = df.iloc[-1,:]["Team"] # replaces the team name from TOT to that player's last team they played for
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"]) # merge dbs (outer merge is used since the mvp db doesn't include all the players so outer merge keeps all the players that won and didn't win mvp )
```

```python
combined
```

```python
combined[["Pts Won", "Pts Max", "Share", "Awards"]] = combined[["Pts Won", "Pts Max", "Share", "Awards"]].fillna(0) # fill NaN values with 0
```
